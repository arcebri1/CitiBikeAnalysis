```python
#Dependencies
import pandas as pd
import datetime
import numpy as np
import math
```

```python
yr2018 = pd.read_csv('Resources/2018/2018_merged.csv')
yr2018.head()
```

```python
yr2018.count()
```

```python
# yr2018.loc[yr2018['gender'] == 1] = 'Male'
# yr2018.head(11)
```

```python
yr2018['gender'].replace([0,1,2],['Unknown','Male', 'Female'],inplace=True)
yr2018.head()
```

```python
yr2018['tripduration'] = yr2018['tripduration']/60
# yr2018.round({'tripduration': 2})
yr2018.head()
```

```python
yr2018 = yr2018.round({'tripduration': 2})
yr2018.head()
```

```python
yr2018['startyear'] = yr2018["starttime"].str[:4]
yr2018['startmonth'] = yr2018['starttime'].str[5:7]
yr2018['startdate'] = yr2018["starttime"].str[:10]
yr2018['starttimeHr'] = yr2018["starttime"].str[11:16]
yr2018['stopyear'] = yr2018["stoptime"].str[:4]
yr2018['stopmonth'] = yr2018['stoptime'].str[5:7]
yr2018['stopdate'] = yr2018["stoptime"].str[:10]
yr2018['stoptimeHr'] = yr2018["stoptime"].str[11:16]
yr2018.head()
```

```python
yr2018.dropna()
```

```python
yr2018.to_csv('Resources/2018/2018_cleaned.csv', index=False)
```
